```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/customer.csv')
```

```python
df.sample(5)
```

```python
df = df.iloc[:,2:]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2)
```

```python
X_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
# Example: Adding a missing category
oe = OrdinalEncoder(categories=[['Poor','Average','Good', 'Excellent'],['School','UG','PG']])
oe.fit(X_train)

# Example: Handling unknown values
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']], handle_unknown='use_encoded_value', unknown_value=-1)
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
```

```python
oe.categories_
```
